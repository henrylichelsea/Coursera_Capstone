```python
# Create a function to get nearby venues from FourSquare API
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
manhattan_venues = getNearbyVenues(names=data_nh['name'],
                                   latitudes=data_nh['lat'],
                                   longitudes=data_nh['lon']
                                  )
```

### Cluster Manhattan neighborhoods based on venue categories

```python
# One hot encoding
train_data = pd.get_dummies(manhattan_venues['Venue Category'], prefix = '', prefix_sep = '')
train_data['Neighbourhood'] = manhattan_venues['Neighbourhood']
train_data = train_data.set_index('Neighbourhood')
```

```python
# Prepare dataset for k-mean clustering
train_data = train_data.groupby(['Neighbourhood']).sum()
train_data.head()
```

```python
# Decide the optimal k for clustering
from sklearn.cluster import KMeans
sse = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, max_iter=1000)
    kmeans.fit(train_data)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 10), sse)
plt.xticks(range(1, 10))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
# Run k-mean clustering using FourSquare category data
from sklearn.cluster import KMeans
kclusters = 5

kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(train_data)

kmeans.labels_[0:10] 
```

```python
# add clustering labels
train_data.insert(0, 'Cluster Labels', kmeans.labels_)
train_data.head()
```

```python
# Merge train_data with manhattan neighbourhood data
manhattan_merged = manhattan_merged.join(train_data, on='Neighbourhood')
manhattan_merged.head()
```

### We consider all Chinese/Asian restaurants are substitutes to Dim Sum restaurants for customers - therefore, we want to see where they are.

```python
# Competition analysis
names = ['Chinese Restaurant', 'Hotpot Restaurant', 'Asian Restaurant', 'Dim Sum Restaurant', 'Taiwanese Restaurant', \
         'Shanghai Restaurant', 'Dumpling Restaurant', 'Cantonese Restaurant', 'Szechuan Restaurant']

competitors = manhattan_venues[manhattan_venues['Venue Category'].isin(names)].copy()
competitors.groupby(['Neighbourhood'])['Venue'].count()
```

```python
# Create a competitor count column for the dataset
cols = ['Neighbourhood', 'borough', 'lon', 'lat', 'Cluster Labels'] + names
dataset_m = manhattan_merged[cols].copy()
dataset_m['Competitor Count'] = dataset_m.iloc[:, -9:-1].sum(axis=1)
dataset_m.head()
```

```python
# Get Walk Score data
walk_score = pd.read_csv(r'/Users/chenli/Documents/GitHub/Coursera_Capstone/Data/Walk_score_by_NYC_Neighbourhood.csv')
walk_score.head()
```

```python
# Left join walk score data with manhattan neighbourhood data using substring matching of neighbourhood column
dataset_mw = dataset_m.copy()
dataset_mw['Walk Score'] = 2.0
dataset_mw['Transit Score'] = 2.0
dataset_mw['Bike Score'] = 2.0
for index, row in dataset_mw.iterrows():
    for a,b in walk_score.iterrows():
        if row['Neighbourhood'] in b['Name']:
            dataset_mw.at[index, 'Walk Score'] = b['Walk Score']
            dataset_mw.at[index, 'Transit Score'] = b['Transit Score']
            dataset_mw.at[index, 'Bike Score'] = b['Bike Score']
        else:
            continue

dataset_mw['Walk Score'] = dataset_mw['Walk Score'].replace(2.0, np.nan)
dataset_mw['Transit Score'] = dataset_mw['Transit Score'].replace(2.0, np.nan)
dataset_mw['Bike Score'] = dataset_mw['Bike Score'].replace(2.0, np.nan)
```
